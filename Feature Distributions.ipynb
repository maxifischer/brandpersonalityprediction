```python
# import plotly components for offline display of diagrams
from IPython.display import display, HTML
import plotly.offline as offline
from plotly.graph_objs import *


# enable latex-mathsymbols in plotly
offline.init_notebook_mode(connected=True)

# The polling here is to ensure that plotly.js has already been loaded before
# setting display alignment in order to avoid a race condition.
display(HTML(
    '<script>'
        'var waitForPlotly = setInterval( function() {'
            'if( typeof(window.Plotly) !== "undefined" ){'
                'MathJax.Hub.Config({ SVG: { font: "STIX-Web" }, displayAlign: "center" });'
                'MathJax.Hub.Queue(["setRenderer", MathJax.Hub, "SVG"]);'
                'clearInterval(waitForPlotly);'
            '}}, 250 );'
    '</script>'
))

# general imports
import pandas as pd
import numpy as np
import datetime
```

```python
import FeatureCreation
```

```python
df = FeatureCreation.get_char_based_lexical_features()
df
```

```python
traces = []
for column in filter(lambda x: x.startswith('letter_'), list(df)):
    trace = Histogram(
        x = df[column],
        name = column
    )
    traces.append(trace)

layout = Layout(
    title = 'Char based features histogram',
    xaxis=dict(
        title='Feature Value',
        type='lin'
    ),
    yaxis=dict(
        title='Count',
        type='log'
    ),
)
fig = dict(data = traces[26:36], layout=layout)
offline.iplot(fig)
```

```python
d = {
    'Column': [],
    'Mean': [],
    'Variance': [],
    'Max': [],
    
}
for column in filter(lambda x: x.startswith('letter_'), list(df)):
    d['Column'].append(column)
    d['Mean'].append(df[column].mean())
    d['Variance'].append(df[column].var())
    d['Max'].append(df[column].max())

keyFigures = pd.DataFrame(d)
keyFigures
```
